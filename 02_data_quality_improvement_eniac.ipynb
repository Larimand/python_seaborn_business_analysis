```python
# orders_cl.csv
url = "https://drive.google.com/file/d/1Tla62vfu__kCqvgypZyVt2S9VuC016yH/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
orders_cl = pd.read_csv(path)

# orderlines_cl.csv
url = "https://drive.google.com/file/d/1OhtkQS2fwOYdzfd-qPh7im35iLc-L9TA/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
orderlines_cl = pd.read_csv(path)

# products_cl.csv
url = "https://drive.google.com/file/d/1s7Lai4NSlsYjGEPg1QSOUJobNYVsZBOJ/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
products_cl = pd.read_csv(path)
```

```python
orders_qu = orders_cl.copy()
orderlines_qu = orderlines_cl.copy()
products_qu = products_cl.copy()
```

 Orders & orderlines

```python
# only keep completed orders
orders_qu['state'].value_counts()
orders_qu = orders_qu.loc[orders_qu['state']=='Completed']
```

```python
# look for all the invalida orderlines
invalid_orderlines = orderlines_qu[~orderlines_qu['sku'].isin(products_qu['sku'])]
# look for the corrupted
corrupt_orderlines = invalid_orderlines['id_order'].unique()
# clean them
clean_orderlines = orderlines_qu.loc[~orderlines_cl['id_order'].isin(corrupt_orderlines)]
# look for invalid orders
invalid_orders = orders_qu.loc[~orders_qu['order_id'].isin(clean_orderlines['id_order'])]
# look for corrupted orders
corrupt_orders = invalid_orders['order_id'].unique()
# clean them
clean_orders = orders_qu[~orders_qu['order_id'].isin(corrupt_orders)]
```

```python
# remove the corresponding orderlines
orderlines_qu['id_order'].nunique()
# merge orders and orderlines
orders_orderlines = pd.merge(orders_qu,
                             orderlines_qu,
                             left_on='order_id',
                             right_on='id_order',
                             how='inner')
```
