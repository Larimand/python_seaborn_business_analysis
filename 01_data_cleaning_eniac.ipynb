Products

```python
# drop duplicates
products_df = products_df.drop_duplicates()
# drop the promo price column (corrupted)
products_df = products_df.drop(columns=['promo_price'])
# drop sku duplicates
products_df.loc[products_df.duplicated(subset=['sku'], keep=False)]
products_df = products_df.drop_duplicates(subset='sku')
# drop none values
products_df = products_df.dropna()
products_df.duplicated(subset=['sku']).sum()
# look for missing values
products_df.isna().sum()
# filling/cleaning empty cells
products_df.loc[products_df['desc'].isna()]
products_df.loc[products_df['desc'].isna(),'desc'] = products_df.loc[products_df['desc'].isna(),'name']
# change data types
mask = products_df['price'].str.count(r'\.') > 1
products_df.loc[mask, 'price'] = products_df.loc[mask, 'price'].str.replace('.', '', n=1)
products_df['price'] = pd.to_numeric(products_df['price'], errors='coerce')
products_df = products_df.astype({'price': 'float64'})
```

```python
products_cl = products_df.copy()
```

Orderlines

```python
# checking orderlines
orderlines_df.info()
orderlines_df.isna()
orderlines_df.duplicated().sum()
```

```python
# change datatypes
# date
orderlines_df["date"] = pd.to_datetime(orderlines_df["date"])
```

```python
# unit price
orderlines_df.loc[orderlines_df['unit_price'].str.count(r'\.')>1,'unit_price'] = \
orderlines_df.loc[orderlines_df['unit_price'].str.count(r'\.')>1,'unit_price'].str.replace('.','',n=1)
orderlines_df["unit_price"] = pd.to_numeric(orderlines_df["unit_price"])
```

```python
orderlines_cl = orderlines_df.copy()
```
